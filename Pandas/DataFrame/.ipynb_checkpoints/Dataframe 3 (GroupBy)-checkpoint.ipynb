```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("imdb-top-1000.csv")
```

```python
movies.head()
```

```python
genres = movies.groupby('Genre')
```

```python
# Applying builtin aggregation function on groupbyobjects
genres.sum()
```

```python
# mean
genres['IMDB_Rating'].sum()
```

```python
# finding the top 3 genres by total earning 
genres['Gross'].sum().sort_values(ascending = False).head(3)
```

```python
# find the genre with the highest avg IMDB rating
genres['IMDB_Rating'].mean().sort_values(ascending = False).head(1)
```

```python
# find director with most popularity
movies[['Director', 'No_of_Votes']].groupby('Director')['No_of_Votes'].sum().sort_values(ascending = False).head(1)
```

```python
# find the highest rated movie of each genre

movies.groupby(['Genre', 'Series_Title'])['IMDB_Rating'].max().reset_index()
```

```python
# find the number of movies done by each actor
movies['Star1'].value_counts()
```

```python
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending = False)
```

```python
# agg method
# passing Dict
genres.sum()
```

```python
# agg method 
# passing dict
genres.agg(
{
    'Runtime': 'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes' : 'sum',
    'Metascore' : 'min',
    'Gross' :'sum'
})
```

```python
# if we want to perform mutiple operation for same column
genres.agg(['min','max', 'sum'])
```

```python
# agg method 
# passing dict
genres.agg(
{
    'Runtime': ['min','max', 'sum'],
    'IMDB_Rating':'mean',
    'No_of_Votes' : ['min','max', 'sum'],
    'Metascore' : 'min',
    'Gross' :'sum'
})
```

```python
# split(apply) combine
# apply -> builtin function
```

```python
# find all the movies which start with Letter A
def foo(group):
    a = group['Series_Title'].str.startswith('A').sum()
    return a
```

```python
genres.apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score
# find all the movies which start with Letter A
def rank_movie(group):
    group['rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalised IMDB rating 
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()
                           )/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normal)
```

```python
# looping on groups
duo = movies.groupby(['Director', 'Star1'])
# size
duo.size()
```

```python
# get_group: Must passed a tuple to perform the operation
duo.get_group(('Aamir Khan', 'Amole Gupte'))
```

```python
# find the most earning actor - > director combo
duo['Gross'].sum().sort_values(ascending = False).head(1)
```

```python
# find the best(in terms of metascore) actor - > genre combo
# we use reset index to convert into a dataframe and the original one was having multi index
movies.groupby(['Star1', 'Genre'])['Metascore'].mean().sort_values(ascending = False).reset_index()
```

```python
# agg on multiple groupby
duo['Gross'].agg(['min', 'max', 'mean'])
```

```python
ipl = pd.read_csv('deliveries.csv')
```

```python
ipl
```

```python
# FInd the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10)
```

```python
# find the batsman with max number of sixes
ipl[ipl['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].count().sort_values(ascending = False).head(10)
```

```python
# find batsman with the most number of 4's and 6's in the last 5 overs
ipl[(ipl['over'] > 15) & ((ipl['batsman_runs'] == 4) | (ipl['batsman_runs'] == 6))].groupby('batsman')['batsman'].count().sort_values(ascending = False).head(10)
```

```python
# find the V kohli's against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```
