# Colored Scatterplots

```python
iris = pd.read_csv("iris (1).csv")
```

```python
iris.head()
```

```python
# need to assign number to each species
# cmap for change
# aplha decide opacity
# c is for grouping for color
# s is for size based on the value

iris['Species_num'] = iris['Species'].replace({'Iris-setosa' : 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species_num'], cmap = 'winter',
           alpha = 0.9)
plt.xlabel("Sepal Length (Cm)")
plt.ylabel("Petal Length(Cm)")
plt.colorbar() # tells the which color represent each category
```

```python
# how to change size of the graph

plt.figure(figsize = (15, 7)) # This is to change the size of the graph

iris['Species_num'] = iris['Species'].replace({'Iris-setosa' : 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species_num'], cmap = 'winter',
           alpha = 0.9)
plt.xlabel("Sepal Length (Cm)")
plt.ylabel("Petal Length(Cm)")
plt.colorbar() # tells the which color represent each category
```

# 3D Surface Plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
z = xx**2 + yy**2
```

```python
z
```

```python
fig = plt.figure(figsize = (16, 8))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx, yy, z , cmap = 'viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize = (16, 8))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx, yy, (np.sin(xx) + np.sin(yy)) , cmap = 'viridis')
fig.colorbar(p)
```

# Contour Plots

```python
fig = plt.figure(figsize = (16, 8))

ax = plt.subplot()
p = ax.contour(xx, yy, (np.sin(xx) + np.sin(yy)) , cmap = 'viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize = (16, 8))

ax = plt.subplot()
p = ax.contourf(xx, yy, (np.sin(xx) + np.sin(yy)) , cmap = 'viridis')
fig.colorbar(p)
```
