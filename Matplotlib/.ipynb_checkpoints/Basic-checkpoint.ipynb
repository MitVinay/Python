```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting a simple function
price = [4000, 4500, 6700, 3000, 5479, 9098]
year = range(2015, 2021)

# first paramter is x and 2nd one is y
plt.plot(year, price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv("sharma-kohli.csv")
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title(" Rohit Sharma Vs Virat Kholi IPL")
plt.xlabel("Runs Scored")
plt.ylabel("Season")
```

```python
# colors(hex) and line(widhth) and marker(size)

plt.plot(batsman['index'], batsman['V Kohli'], color ='green')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black')

plt.title(" Rohit Sharma Vs Virat Kholi IPL")
plt.xlabel("Runs Scored")
plt.ylabel("Season")
```

```python
# line style and width

plt.plot(batsman['index'], batsman['V Kohli'], color ='green', linestyle = 'solid', linewidth = 2)
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashed', linewidth = 3)

plt.title(" Rohit Sharma Vs Virat Kholi IPL")
plt.xlabel("Runs Scored")
plt.ylabel("Season")
```

```python
# adding markers: the marker is the coordinate and marker size


plt.plot(batsman['index'], batsman['V Kohli'], color ='green', linestyle = 'solid', linewidth = 2, marker = 'o')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashed', linewidth = 3,
         marker = '>', markersize = 10)

plt.title(" Rohit Sharma Vs Virat Kholi IPL")
plt.xlabel("Runs Scored")
plt.ylabel("Season")
```

```python
# legend

plt.plot(batsman['index'], batsman['V Kohli'], color ='green', linestyle = 'solid', linewidth = 2, marker = 'o',
        label = 'V Kohli')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashed', linewidth = 3,
         marker = '>', markersize = 10, label = 'RG Sharma')

plt.title(" Rohit Sharma Vs Virat Kholi IPL")
plt.xlabel("Runs Scored")
plt.ylabel("Season")
# plt.legend() " Default paramter is best, which sets the legend based on the space"
# plt.legend(loc = "upper right")
plt.legend(loc = 'best')
```

```python
# limiting axis
# plotting a simple function
price = [4000, 4500, 6700, 3000, 5479, 9098, 4500000]
year = range(2015, 2022)

# first paramter is x and 2nd one is y
plt.plot(year, price)
plt.ylim(0, 10000)
plt.xlim(2015, 2020)
```

```python
# grid

plt.plot(batsman['index'], batsman['V Kohli'], color ='green', linestyle = 'solid', linewidth = 2, marker = 'o',
        label = 'V Kohli')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashed', linewidth = 3,
         marker = '>', markersize = 10, label = 'RG Sharma')

plt.title(" Rohit Sharma Vs Virat Kholi IPL")
plt.xlabel("Runs Scored")
plt.ylabel("Season")
# plt.legend() " Default paramter is best, which sets the legend based on the space"
# plt.legend(loc = "upper right")
plt.legend(loc = 'best')
plt.grid()
```

```python
# show: it is the main function to show the graph , other IDE like pycharm we need to specifically mentioned
# the use of show function.

plt.plot(batsman['index'], batsman['V Kohli'], color ='green', linestyle = 'solid', linewidth = 2, marker = 'o',
        label = 'V Kohli')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashed', linewidth = 3,
         marker = '>', markersize = 10, label = 'RG Sharma')

plt.title(" Rohit Sharma Vs Virat Kholi IPL")
plt.xlabel("Runs Scored")
plt.ylabel("Season")
# plt.legend() " Default paramter is best, which sets the legend based on the space"
# plt.legend(loc = "upper right")
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

## Scatter Plot

- Bi-variate Analysis
- Numerical VS Numerical
- Use case - Finding Correlation


```python
# plt.scatter simple function
x = np.linspace(-10, 10, 50)
y = 10*x + 3 + np.random.randint(0, 300, 50)
```

```python
plt.scatter(x, y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv("batter.csv")
```

```python
plt.scatter(df['avg'].head(50), df['strike_rate'].head(50), color = "green", marker = '+')
plt.title("Avg and Strike rate analysis of top 50")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
```

```python
# size based on any parameter
tips = sns.load_dataset('tips')
plt.scatter(tips['total_bill'], tips['tip'], s = tips['size']*10, color = 'green')
```

```python
# scatterplot using plt.plot: but cannot using s functionality
plt.plot(tips['total_bill'], tips['tip'],'o', color = 'green')
```

## Bar chart

- Bivariate analysis
- Numerical VS Categorical
- Use Case -  Aggregate analysis of Groups

```python
# simple bar chart
children = [5, 10, 15, 3, 9]
colors = ['green', 'red', 'blue', 'pink', 'yellow']
```

```python
plt.bar(colors, children, color = 'black')
```

```python
# bar chart using data: horizontal
plt.barh(colors, children, color = 'black')
```

```python
# color and label
df = pd.read_csv("batsman_season_record.csv")
```

```python
df
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])
```

```python
# better way to project all year 
plt.bar(np.arange(df.shape[0]) - 0.3, df['2015'], width=0.285)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.285)
plt.bar(np.arange(df.shape[0]) + 0.3, df['2017'], width=0.285)

# changing existing labels
plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# rotation parameter
plt.bar(colors, children, color = 'black')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# stacked bar chart
plt.bar(df['batsman'], df['2017'], label = '2017')
plt.bar(df['batsman'], df['2016'], bottom = df['2017'], label = '2016') 
plt.bar(df['batsman'], df['2015'], bottom = df['2016'] + df['2017'], label = '2015')

plt.legend()
```

## Pie Chart

- Uni-variate/ Bivariate
- Categorical vs numerical
- Use case- to find cotribution on standard scale

```python
data = [28, 50]
# 28/(28 + 50)
subjects = ['eng' , 'hindi']
plt.pie(data, labels= subjects)
plt.show()
```

```python
df = pd.read_csv("gayle-175.csv")
```

```python
plt.pie(df['batsman_runs'], labels= df['batsman'])
plt.show()
```

```python
# percentage contrubution
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct = '%0.01f%%')
plt.show()
```

```python
# colors contrubution
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct = '%0.01f%%', colors=['blue', 'green', 'yellow', 'pink',
                                                                                'brown', 'red'])
plt.show()
```

```python
# explode shadow: we want to extrcr  partiuclar pie
# colors contrubution
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct = '%0.01f%%',
        colors=['blue', 'green', 'yellow', 'pink','brown', 'red'], explode = [0.1,0,0,0,0,0])
plt.show()
```

```python
# explode shadow: we want to extrcr  partiuclar pie
# colors contrubution
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct = '%0.01f%%',
        colors=['blue', 'green', 'yellow', 'pink','brown', 'red'], explode = [0.1,0,0,0,0,0],
       shadow=True)
plt.show()
```
