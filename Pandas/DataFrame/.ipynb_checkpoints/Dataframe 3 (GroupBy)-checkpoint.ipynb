```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("imdb-top-1000.csv")
```

```python
movies.head()
```

```python
genres = movies.groupby('Genre')
```

```python
# Applying builtin aggregation function on groupbyobjects
genres.sum()
```

```python
# mean
genres['IMDB_Rating'].sum()
```

```python
# finding the top 3 genres by total earning 
genres['Gross'].sum().sort_values(ascending = False).head(3)
```

```python
# find the genre with the highest avg IMDB rating
genres['IMDB_Rating'].mean().sort_values(ascending = False).head(1)
```

```python
# find director with most popularity
movies[['Director', 'No_of_Votes']].groupby('Director')['No_of_Votes'].sum().sort_values(ascending = False).head(1)
```

```python
# find the highest rated movie of each genre

movies.groupby(['Genre', 'Series_Title'])['IMDB_Rating'].max().reset_index()
```

```python
# find the number of movies done by each actor
movies['Star1'].value_counts()
```

```python
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending = False)
```

```python
# GroupBy Attribute and Methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> vs filtering
# groups
# describe
# sample
# nunique
```

```python
# find the number of genres
len(movies.groupby('Genre'))
```

```python
movies['Genre'].nunique()
```

```python
# Number of movies in each genre
movies.groupby('Genre').size()
```

```python
# First movie of each genre
genres.first()
```

```python
# last movie of each genre
genres.last()
```

```python
# nth group
genres.nth(2)
```

```python
# get_group: show all horror  movies

genres.get_group('Horror')
```

```python
# groups: gives the dictionary, and key is group andvalue will be list of indexbb
genres.groups
```

```python
# describe: given describe data for all the groups
genres.describe()
```

```python
# sample gives random movie from each genre
genres.sample()
```

```python
# nunique 
genres.nunique()
```

```python
# agg method
# passing Dict
genres.sum()
```

```python
# agg method 
# passing dict
genres.agg(
{
    'Runtime': 'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes' : 'sum',
    'Metascore' : 'min',
    'Gross' :'sum'
})
```

```python
# if we want to perform mutiple operation for same column
genres.agg(['min','max', 'sum'])
```

```python
# agg method 
# passing dict
genres.agg(
{
    'Runtime': ['min','max', 'sum'],
    'IMDB_Rating':'mean',
    'No_of_Votes' : ['min','max', 'sum'],
    'Metascore' : 'min',
    'Gross' :'sum'
})
```

```python
# looping on groups: it wiill give two thing one group object and other is data
for group, data in genres:
    print(type(group), type(data))
    print(data)
```
