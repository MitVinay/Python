```python
ipl = pd.read_csv("ipl-matches.csv")
```

```python
# find which player has won most pom -> finals and qualifers
# need str to work with string fuction
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts().head(10)
```

```python
# Toss Decision plot: On series we can use plot function
ipl['TossDecision'].value_counts().plot(kind = 'pie')
```

```python
## How many matches each team has played
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending= False)
```

### sort_values( Series and DataFrame) -> ascending -> na_position -> inplace -> multiple cols

```python
movies = pd.read_csv("movies.csv")
```

```python
# Sorting the data frame on the basis of one single column
movies.sort_values('title_x', ascending= False)
```

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)

students
```

```python
# what happens when we apply sort_values function to column having nan values
students.sort_values('name')
# if we notice the missing values will go in the last or top based on the na_position. default is last
students.sort_values('name', na_position='first', ascending=False)
```

```python
# to permanantly save to  the orignal data frame
students.sort_values('name', na_position='first', ascending=False, inplace=True)
```

```python
students
```

```python
# Nested Sorting
movies.sort_values(['year_of_release', 'title_x'])

# Nested Sorting: Ascending descending combination
movies.sort_values(['year_of_release', 'title_x'], ascending=[True, False])
```
