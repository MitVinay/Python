## Selecting Rows From a DataFrame

- iloc : searches using index position
- loc : searches using index labels

```python
# Using Dictionary
student_dic1 = { "Name" : ['Vinay', 'Vineet', 'Rohan', 'Vatsal', 'Harsh'],
    'IQ' : [100, 90, 120, 1, 1], "Marks" : [80, 70, 100, 1, 1], "Package" : [10, 7, 12, 1, 1] }
```

```python
student2 = pd.DataFrame(student_dic1)
```

```python
student2.set_index("Name", inplace= True)
```

```python
student2
```

```python
# single row: pulling first row
movies.iloc[0]
```

```python
type(movies.iloc[0])
```

```python
# Multiple Row
movies.iloc[0:5]
```

```python
movies.iloc[0:5:2]
```

```python
# Fancy indexing
movies.iloc[[1, 4, 5]]
```

```python
# loc
student2.loc['Vinay']
```

```python
student2.loc[['Vinay', "Vineet"]]
```

```python
student2.loc['Vinay': "Vatsal"]
```

*Major difference between iloc and loc is that in iloc the last index is not included, Where as in loc the last index is included. As we can see in the above.*
