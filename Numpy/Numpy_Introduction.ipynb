# Numpy Operation

```python
"""In numpy operation, same index item are considered for the operation """
arr1 = np.arange(1, 11)
arr2 = np.arange(10, 20)
```

```python
print(arr1 - arr2)
print(arr1 + arr2)
print(arr1 * arr2)
print(arr2 / arr1)
```

```python
arr1 > 2
```

```python
arr3 = np.arange(6).reshape(2, 3)
arr4 = np.arange(6, 12).reshape(3, 2)
```

```python
arr3.dot(arr4)
```

```python
arr4.max()
```

```python
arr4.min()
```

```python
# column wise minimum
print(arr4.min(axis = 0))

# row wise minimum
print(arr4.min(axis = 1))
```

```python
# Sum
arr4.sum()

# column wise sum
print(arr4.sum(axis = 0))

# row wise sum
print(arr4.sum(axis = 1))
```

```python
# Mean
print(arr4.mean())

# column wise mean
print(arr4.mean(axis = 0))

# row wise mean
print(arr4.mean(axis = 1))
```

```python
# Standard Deviation
print(arr4.std())

# column wise Standard Deviation
print(arr4.std(axis = 0))

# row wise Standard Deviation
print(arr4.std(axis = 1))
```

```python
# Column wise Median

print(np.median(arr4, axis = 0))
# Row wise Median

print(np.median(arr4, axis = 1))
```

```python
# Sin Function

print(np.sin(arr4))
```

```python
# Exponent 
np.exp(arr3 )
```
