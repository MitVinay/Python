```python
import pandas as pd
import numpy as np
```

```python
# reading CSV file f
courses = pd.read_csv("courses.csv")
students = pd.read_csv("students.csv")
nov = pd.read_csv("reg-month1.csv")
dec = pd.read_csv("reg-month2.csv")
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
```

```python
# Vertical stacking without taking the original index due to repitation
regs = pd.concat([nov, dec], ignore_index = True)
```

```python
# Mutli index data frame
multi = pd.concat([nov, dec], keys = ['Nov', 'Dec'])
```

```python
multi.loc['Nov']
```

```python
# It will be passed in tuple
multi.loc[('Nov', 0)]
```

```python
# horizontal stacking
pd.concat([nov, dec], axis = 1)
```

```python
# inner join
students.merge(regs, how = 'inner', on = 'student_id')
```

```python
# Left outer join
students.merge(regs, how = 'left', on = 'student_id')
```
